```python
# =============================================================================
# STEP 5: CREATE FEATURE EXTRACTION FUNCTION
# =============================================================================
def extract_discharge_features(battery_data, battery_name):
    """
    Extract features from all discharge cycles in a battery file
    """
    features_list = []
    capacities = []

    num_cycles = battery_data['cycle'][0, 0].shape[1]

    for cycle_idx in range(num_cycles):
        cycle_type = battery_data['cycle'][0, 0]['type'][0, cycle_idx][0]

        # Only process discharge cycles
        if 'discharge' in cycle_type.lower():
            cycle_data = battery_data['cycle'][0, 0]['data'][0, cycle_idx]

            # Extract measurements
            voltage = cycle_data['Voltage_measured'][0, 0].flatten()
            current = cycle_data['Current_measured'][0, 0].flatten()
            temperature = cycle_data['Temperature_measured'][0, 0].flatten()
            time = cycle_data['Time'][0, 0].flatten()
            capacity = cycle_data['Capacity'][0, 0].flatten()[0]  # Single value

            # Skip cycles with insufficient data
            if len(voltage) < 10:
                continue

            # Calculate features
            features = {
                'battery': battery_name,
                'cycle': cycle_idx + 1,
                'capacity': capacity,
                'voltage_mean': np.mean(voltage),
                'voltage_std': np.std(voltage),
                'voltage_min': np.min(voltage),
                'voltage_max': np.max(voltage),
                'current_mean': np.mean(current),
                'current_std': np.std(current),
                'temperature_mean': np.mean(temperature),
                'temperature_max': np.max(temperature),
                'discharge_time': time[-1] - time[0],  # Total discharge duration
                'num_measurements': len(voltage)
            }

            features_list.append(features)
            capacities.append(capacity)

    return features_list, capacities

print("‚úÖ FEATURE EXTRACTION FUNCTION CREATED!")
print("üîß This function will extract from each discharge cycle:")
print("   ‚Ä¢ Voltage statistics (mean, std, min, max)")
print("   ‚Ä¢ Current statistics (mean, std)")
print("   ‚Ä¢ Temperature statistics (mean, max)")
print("   ‚Ä¢ Discharge time and capacity")
```

```python
# =============================================================================
# STEP 8: DATA CLEANING AND CATEGORICAL LABELS
# =============================================================================
print("üßπ CLEANING DATA AND CREATING CATEGORICAL LABELS...")

# Remove any cycles with invalid capacity (0 or negative)
initial_count = len(features_df)
features_df = features_df[features_df['capacity'] > 0.01]  # Remove near-zero capacities
print(f"üìä Removed {initial_count - len(features_df)} cycles with invalid capacity")

# Calculate SOH for each battery
print("\nüîã CALCULATING STATE OF HEALTH (SOH)...")

# Group by battery and calculate max capacity (100% SOH reference)
battery_max_capacity = features_df.groupby('battery')['capacity'].max().to_dict()

# Calculate SOH for each cycle (current capacity / max capacity * 100)
features_df['soh'] = features_df.apply(
    lambda row: (row['capacity'] / battery_max_capacity[row['battery']]) * 100,
    axis=1
)

print(f"‚úÖ SOH calculation complete!")
print(f"üìà SOH range: {features_df['soh'].min():.2f}% to {features_df['soh'].max():.2f}%")

# CREATE CATEGORICAL LABELS (Good/Medium/Bad)
print("\nüéØ CREATING CATEGORICAL LABELS...")

def categorize_soh(soh):
    if soh >= 80:
        return 'good'
    elif soh >= 60:
        return 'medium'
    else:
        return 'bad'

features_df['health_label'] = features_df['soh'].apply(categorize_soh)

# Show label distribution
label_counts = features_df['health_label'].value_counts()
print(f"üìä HEALTH LABEL DISTRIBUTION:")
print(f"   ‚Ä¢ Good:   {label_counts.get('good', 0):>4} cycles")
print(f"   ‚Ä¢ Medium: {label_counts.get('medium', 0):>4} cycles")
print(f"   ‚Ä¢ Bad:    {label_counts.get('bad', 0):>4} cycles")

# Show SOH ranges for each category
print(f"\nüìà SOH CATEGORY RANGES:")
print(f"   ‚Ä¢ Good:   SOH ‚â• 80%")
print(f"   ‚Ä¢ Medium: 60% ‚â§ SOH < 80%")
print(f"   ‚Ä¢ Bad:    SOH < 60%")

# Show examples
print(f"\nüìã SAMPLE DATA WITH LABELS:")
sample_data = features_df[['battery', 'cycle', 'capacity', 'soh', 'health_label', 'discharge_time']].head(10)
print(sample_data)

# Final dataset info
print(f"\nüéØ FINAL DATASET:")
print(f"   ‚Ä¢ Total cycles: {len(features_df):,}")
print(f"   ‚Ä¢ Total batteries: {features_df['battery'].nunique()}")
print(f"   ‚Ä¢ Features per cycle: {features_df.shape[1]}")
print(f"   ‚Ä¢ Time feature: discharge_time (already included)")
```

```python
# =============================================================================
# STEP 9: PREPARE DATA FOR GRU TRAINING
# =============================================================================
print("üìä PREPARING DATA FOR GRU TRAINING...")
print("=" * 60)

# Select features for training
feature_columns = ['voltage_mean', 'voltage_std', 'current_mean', 'current_std',
                   'temperature_mean', 'temperature_max', 'discharge_time', 'num_measurements']

print(f"üéØ SELECTED FEATURES: {feature_columns}")

# Prepare feature matrix X and labels y
X = features_df[feature_columns].values
y = features_df['health_label'].values

print(f"üì¶ DATA SHAPES:")
print(f"   ‚Ä¢ X (features): {X.shape}")
print(f"   ‚Ä¢ y (labels): {y.shape}")

# Encode labels to numerical values
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print(f"üî§ LABEL ENCODING:")
for i, label in enumerate(label_encoder.classes_):
    print(f"   ‚Ä¢ {label} ‚Üí {i}")

print(f"   ‚Ä¢ Encoded y: {y_encoded.shape}")

# Scale features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(f"üìè FEATURE SCALING:")
print(f"   ‚Ä¢ Scaled X: {X_scaled.shape}")
print(f"   ‚Ä¢ Feature means: {scaler.mean_.round(3)}")
print(f"   ‚Ä¢ Feature stds: {scaler.scale_.round(3)}")

print("=" * 60)
print("‚úÖ DATA PREPARATION COMPLETE!")
```

```python
# =============================================================================
# STEP 10: CREATE SEQUENCES FOR GRU AND SPLIT DATA
# =============================================================================
print("üîÑ CREATING SEQUENCES FOR GRU TRAINING...")
print("=" * 60)

# Create sequences for time-series prediction
def create_sequences(features, labels, sequence_length=10):
    X_sequences = []
    y_sequences = []

    # Group by battery to maintain temporal order
    for battery in features_df['battery'].unique():
        battery_mask = features_df['battery'] == battery
        battery_features = features[battery_mask]
        battery_labels = labels[battery_mask]

        # Create sequences for this battery
        for i in range(len(battery_features) - sequence_length):
            X_sequences.append(battery_features[i:i + sequence_length])
            y_sequences.append(battery_labels[i + sequence_length])

    return np.array(X_sequences), np.array(y_sequences)

# Create sequences
sequence_length = 10
X_sequences, y_sequences = create_sequences(X_scaled, y_encoded, sequence_length)

print(f"üìä SEQUENCE DATA SHAPES:")
print(f"   ‚Ä¢ X_sequences: {X_sequences.shape} (sequences, time_steps, features)")
print(f"   ‚Ä¢ y_sequences: {y_sequences.shape} (target labels)")
print(f"   ‚Ä¢ Sequence length: {sequence_length} time steps")
print(f"   ‚Ä¢ Features per time step: {X_sequences.shape[2]}")

# Split data into train/validation/test
from sklearn.model_selection import train_test_split

# First split: separate test set
X_temp, X_test, y_temp, y_test = train_test_split(
    X_sequences, y_sequences, test_size=0.2, random_state=42, stratify=y_sequences
)

# Second split: separate validation set
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.2, random_state=42, stratify=y_temp
)

print(f"\nüìÇ DATA SPLIT:")
print(f"   ‚Ä¢ Training set: {X_train.shape[0]:,} sequences")
print(f"   ‚Ä¢ Validation set: {X_val.shape[0]:,} sequences")
print(f"   ‚Ä¢ Test set: {X_test.shape[0]:,} sequences")
print(f"   ‚Ä¢ Total sequences: {X_sequences.shape[0]:,}")

# Check label distribution in each set
print(f"\nüéØ LABEL DISTRIBUTION:")
for set_name, y_data in [("Training", y_train), ("Validation", y_val), ("Test", y_test)]:
    unique, counts = np.unique(y_data, return_counts=True)
    print(f"   ‚Ä¢ {set_name:>11}: ", end="")
    for label, count in zip(unique, counts):
        label_name = label_encoder.inverse_transform([label])[0]
        percentage = (count / len(y_data)) * 100
        print(f"{label_name}: {count} ({percentage:.1f}%)", end="  ")
    print()

print("=" * 60)
print("‚úÖ SEQUENCE CREATION AND DATA SPLIT COMPLETE!")
```

```python
# =============================================================================
# STEP 11: BUILD AND COMPILE GRU MODEL
# =============================================================================
print("üß† BUILDING GRU MODEL FOR BATTERY HEALTH PREDICTION...")
print("=" * 60)

# Clear any previous sessions
tf.keras.backend.clear_session()

# Build GRU model
model = Sequential([
    # First GRU layer
    GRU(64, activation='tanh', return_sequences=True,
        input_shape=(sequence_length, X_train.shape[2]),
        dropout=0.2, recurrent_dropout=0.2),

    # Second GRU layer
    GRU(32, activation='tanh', return_sequences=False,
        dropout=0.2, recurrent_dropout=0.2),

    # Dense layers
    Dense(32, activation='relu'),
    Dropout(0.3),

    Dense(16, activation='relu'),
    Dropout(0.2),

    # Output layer (3 classes: bad, good, medium)
    Dense(3, activation='softmax')
])

# Compile the model
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy', 'sparse_categorical_accuracy']
)

print("‚úÖ MODEL ARCHITECTURE:")
model.summary()

print(f"\nüéØ MODEL CONFIGURATION:")
print(f"   ‚Ä¢ Input shape: ({sequence_length}, {X_train.shape[2]})")
print(f"   ‚Ä¢ Output: 3 classes ({', '.join(label_encoder.classes_)})")
print(f"   ‚Ä¢ Learning rate: 0.001")
print(f"   ‚Ä¢ Optimizer: Adam")
print(f"   ‚Ä¢ Loss: Sparse Categorical Crossentropy")

print("=" * 60)
print("‚úÖ GRU MODEL BUILT AND COMPILED SUCCESSFULLY!")
```

```python
# =============================================================================
# STEP 12: TRAIN THE GRU MODEL
# =============================================================================
print("üöÄ TRAINING GRU MODEL ON BATTERY DATA...")
print("=" * 60)

# Set up callbacks
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=15,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=10,
    min_lr=0.0001,
    verbose=1
)

# Train the model
print("üèãÔ∏è‚Äç‚ôÇÔ∏è STARTING TRAINING...")
history = model.fit(
    X_train, y_train,
    batch_size=32,
    epochs=100,
    validation_data=(X_val, y_val),
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)

print("=" * 60)
print("‚úÖ GRU MODEL TRAINING COMPLETED!")

# Show training summary
final_epoch = len(history.history['loss'])
print(f"\nüìä TRAINING SUMMARY:")
print(f"   ‚Ä¢ Epochs trained: {final_epoch}")
print(f"   ‚Ä¢ Final training accuracy: {history.history['accuracy'][-1]:.4f}")
print(f"   ‚Ä¢ Final validation accuracy: {history.history['val_accuracy'][-1]:.4f}")
print(f"   ‚Ä¢ Final training loss: {history.history['loss'][-1]:.4f}")
print(f"   ‚Ä¢ Final validation loss: {history.history['val_loss'][-1]:.4f}")
```

```python
# =============================================================================
# STEP 13: EVALUATE MODEL ON TEST SET
# =============================================================================
print("üß™ EVALUATING MODEL ON TEST SET...")
print("=" * 60)

# Evaluate on test set
test_loss, test_accuracy, test_sparse_accuracy = model.evaluate(X_test, y_test, verbose=0)

print(f"üéØ TEST SET PERFORMANCE:")
print(f"   ‚Ä¢ Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
print(f"   ‚Ä¢ Test Loss: {test_loss:.4f}")
print(f"   ‚Ä¢ Sparse Categorical Accuracy: {test_sparse_accuracy:.4f}")

# Make predictions
y_pred_proba = model.predict(X_test, verbose=0)
y_pred = np.argmax(y_pred_proba, axis=1)

# Calculate additional metrics
from sklearn.metrics import classification_report, confusion_matrix

print(f"\nüìä DETAILED CLASSIFICATION REPORT:")
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(f"üìà CONFUSION MATRIX:")
print(f"   Actual\\Predicted  {label_encoder.classes_[0]:>6}  {label_encoder.classes_[1]:>6}  {label_encoder.classes_[2]:>6}")
for i, true_label in enumerate(label_encoder.classes_):
    print(f"   {true_label:>14} {cm[i][0]:>6} {cm[i][1]:>6} {cm[i][2]:>6}")

# Calculate overall performance
from sklearn.metrics import accuracy_score

overall_accuracy = accuracy_score(y_test, y_pred)
print(f"\n‚úÖ OVERALL TEST PERFORMANCE:")
print(f"   ‚Ä¢ Overall Accuracy: {overall_accuracy:.4f} ({overall_accuracy*100:.2f}%)")
print(f"   ‚Ä¢ Total Test Samples: {len(y_test)}")
print(f"   ‚Ä¢ Correct Predictions: {np.sum(y_pred == y_test)}")
print(f"   ‚Ä¢ Wrong Predictions: {np.sum(y_pred != y_test)}")

print("=" * 60)
print("‚úÖ MODEL EVALUATION COMPLETED!")
```

```python
# =============================================================================
# STEP 19: FIX SEQUENCE CREATION - PROPER TIME ALIGNMENT
# =============================================================================
print("üîÑ FIXING SEQUENCE CREATION WITH PROPER TIME ALIGNMENT...")
print("=" * 60)

def create_sequences_correct(features_df, features, labels, sequence_length=10):
    """
    CORRECT sequence creation: Use past sequences to predict future health
    """
    X_sequences = []
    y_sequences = []
    battery_names = []

    # Group by battery to maintain temporal order
    for battery in features_df['battery'].unique():
        battery_mask = features_df['battery'] == battery
        battery_features = features[battery_mask]
        battery_labels = labels[battery_mask]
        battery_cycles = features_df[battery_mask]['cycle'].values

        # Create sequences for this battery
        for i in range(len(battery_features) - sequence_length):
            # Use sequence from cycles [i, i+sequence_length-1]
            # to predict health at cycle i+sequence_length
            X_sequences.append(battery_features[i:i + sequence_length])
            y_sequences.append(battery_labels[i + sequence_length])
            battery_names.append(battery)

    return np.array(X_sequences), np.array(y_sequences), battery_names

print("üîÑ RECREATING SEQUENCES WITH CORRECT TIME ALIGNMENT...")

# Get the original unscaled features for correct sequence creation
original_features = features_df[feature_columns].values
original_labels = y_encoded

# Create sequences with correct alignment
X_sequences_correct, y_sequences_correct, sequence_batteries = create_sequences_correct(
    features_df, original_features, original_labels, sequence_length
)

print(f"‚úÖ CORRECT SEQUENCE SHAPES:")
print(f"   ‚Ä¢ X_sequences: {X_sequences_correct.shape}")
print(f"   ‚Ä¢ y_sequences: {y_sequences_correct.shape}")

# Scale the features AFTER sequence creation (important!)
scaler_correct = StandardScaler()
# Fit on all sequence data (flattened)
X_sequences_flat = X_sequences_correct.reshape(-1, X_sequences_correct.shape[2])
scaler_correct.fit(X_sequences_flat)

# Scale the sequences
X_sequences_scaled = np.array([scaler_correct.transform(seq) for seq in X_sequences_correct])

print(f"   ‚Ä¢ X_sequences_scaled: {X_sequences_scaled.shape}")

# Split the corrected data
X_temp_c, X_test_c, y_temp_c, y_test_c = train_test_split(
    X_sequences_scaled, y_sequences_correct, test_size=0.2, random_state=42, stratify=y_sequences_correct
)

X_train_c, X_val_c, y_train_c, y_val_c = train_test_split(
    X_temp_c, y_temp_c, test_size=0.2, random_state=42, stratify=y_temp_c
)

print(f"\nüìÇ CORRECTED DATA SPLIT:")
print(f"   ‚Ä¢ Training: {X_train_c.shape[0]:,} sequences")
print(f"   ‚Ä¢ Validation: {X_val_c.shape[0]:,} sequences")
print(f"   ‚Ä¢ Test: {X_test_c.shape[0]:,} sequences")

# Verify label distribution
print(f"\nüéØ CORRECTED LABEL DISTRIBUTION:")
for set_name, X_data, y_data in [("Training", X_train_c, y_train_c),
                                ("Validation", X_val_c, y_val_c),
                                ("Test", X_test_c, y_test_c)]:
    unique, counts = np.unique(y_data, return_counts=True)
    print(f"   ‚Ä¢ {set_name:>11}: ", end="")
    for label, count in zip(unique, counts):
        label_name = label_encoder.inverse_transform([label])[0]
        percentage = (count / len(y_data)) * 100
        print(f"{label_name}: {count} ({percentage:.1f}%)", end="  ")
    print()

print("=" * 60)
print("‚úÖ SEQUENCES CORRECTED! READY FOR RETRAINING!")
```

```python
# =============================================================================
# STEP 20: RETRAIN MODEL WITH CORRECTED SEQUENCES
# =============================================================================
print("üöÄ RETRAINING MODEL WITH CORRECTED SEQUENCES...")
print("=" * 60)

# Clear previous model
tf.keras.backend.clear_session()

# Build the same model architecture
model_corrected = Sequential([
    GRU(64, activation='tanh', return_sequences=True,
        input_shape=(sequence_length, X_train_c.shape[2]),
        dropout=0.2, recurrent_dropout=0.2),
    GRU(32, activation='tanh', return_sequences=False,
        dropout=0.2, recurrent_dropout=0.2),
    Dense(32, activation='relu'),
    Dropout(0.3),
    Dense(16, activation='relu'),
    Dropout(0.2),
    Dense(3, activation='softmax')
])

# Compile the model
model_corrected.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("‚úÖ MODEL ARCHITECTURE (CORRECTED):")
model_corrected.summary()

# Calculate class weights for the corrected data
class_weights_corrected = compute_class_weight(
    'balanced',
    classes=np.unique(y_train_c),
    y=y_train_c
)
class_weight_dict_corrected = {i: weight for i, weight in enumerate(class_weights_corrected)}

print(f"\nüéØ CLASS WEIGHTS (CORRECTED):")
for class_idx, weight in class_weight_dict_corrected.items():
    class_name = label_encoder.inverse_transform([class_idx])[0]
    print(f"   ‚Ä¢ {class_name}: {weight:.2f}")

# Train with corrected sequences
print(f"\nüèãÔ∏è‚Äç‚ôÇÔ∏è TRAINING WITH CORRECTED SEQUENCES...")
history_corrected = model_corrected.fit(
    X_train_c, y_train_c,
    batch_size=32,
    epochs=50,
    validation_data=(X_val_c, y_val_c),
    class_weight=class_weight_dict_corrected,
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)

print("=" * 60)
print("‚úÖ CORRECTED MODEL TRAINING COMPLETED!")
```

```python
# =============================================================================
# STEP 26: ERROR ANALYSIS PLOTS ONLY
# =============================================================================
print("üìä PLOTTING PREDICTION ERRORS ANALYSIS")
print("=" * 60)

import matplotlib.pyplot as plt

# Get all predictions
y_pred_all = model_corrected.predict(X_test_c, verbose=0)
y_pred_classes = np.argmax(y_pred_all, axis=1)

# Create error mask
errors_mask = y_pred_classes != y_test_c
correct_mask = y_pred_classes == y_test_c

print(f"üìà ERROR ANALYSIS:")
print(f"   ‚Ä¢ Total test samples: {len(y_test_c)}")
print(f"   ‚Ä¢ Errors: {np.sum(errors_mask)} ({np.sum(errors_mask)/len(y_test_c)*100:.1f}%)")
print(f"   ‚Ä¢ Correct: {np.sum(correct_mask)} ({np.sum(correct_mask)/len(y_test_c)*100:.1f}%)")

# Plot 1: Error distribution by actual class
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
error_by_class = []
for class_idx in range(3):
    class_mask = y_test_c == class_idx
    class_errors = np.sum((y_test_c == class_idx) & errors_mask)
    class_total = np.sum(class_mask)
    error_rate = class_errors / class_total * 100
    error_by_class.append(error_rate)

class_names = label_encoder.classes_
plt.bar(class_names, error_by_class, color=['red', 'green', 'orange'])
plt.title('Error Rate by Actual Class')
plt.ylabel('Error Rate (%)')
for i, v in enumerate(error_by_class):
    plt.text(i, v + 1, f'{v:.1f}%', ha='center')

# Plot 2: Confidence distribution for errors vs correct
plt.subplot(2, 3, 2)
confidences = np.max(y_pred_all, axis=1)
plt.hist([confidences[correct_mask], confidences[errors_mask]],
         bins=20, alpha=0.7, label=['Correct', 'Errors'], color=['green', 'red'])
plt.title('Confidence Distribution: Correct vs Errors')
plt.xlabel('Prediction Confidence')
plt.ylabel('Frequency')
plt.legend()

# Plot 3: Error pattern - what classes are confused with what?
plt.subplot(2, 3, 3)
confusion = np.zeros((3, 3))
for actual in range(3):
    for predicted in range(3):
        if actual != predicted:
            count = np.sum((y_test_c == actual) & (y_pred_classes == predicted))
            confusion[actual, predicted] = count

plt.imshow(confusion, cmap='Reds', interpolation='nearest')
plt.colorbar()
plt.xticks(range(3), label_encoder.classes_, rotation=45)
plt.yticks(range(3), label_encoder.classes_)
plt.title('Error Confusion Pattern')
for i in range(3):
    for j in range(3):
        if i != j and confusion[i, j] > 0:
            plt.text(j, i, f'{int(confusion[i, j])}',
                    ha='center', va='center', fontweight='bold')

# Plot 4: Feature importance for errors
plt.subplot(2, 3, 4)
# Calculate which features have largest differences between errors and correct
feature_differences = []
for feature_idx in range(X_test_c.shape[2]):
    error_feature_mean = np.abs(X_test_c[errors_mask, :, feature_idx]).mean()
    correct_feature_mean = np.abs(X_test_c[correct_mask, :, feature_idx]).mean()
    difference = error_feature_mean - correct_feature_mean
    feature_differences.append(difference)

feature_names = ['Volt_mean', 'Volt_std', 'Curr_mean', 'Curr_std',
                 'Temp_mean', 'Temp_max', 'Time', 'Samples']
plt.barh(feature_names, feature_differences, color='purple')
plt.title('Feature Differences: Errors vs Correct')
plt.xlabel('Mean Absolute Difference')

# Plot 5: Error distribution across batteries
plt.subplot(2, 3, 5)
battery_errors = {}
for battery in np.unique(sequence_batteries):
    battery_mask = np.array([battery == b for b in sequence_batteries[:len(X_test_c)]])
    if np.sum(battery_mask) > 0:
        battery_error_rate = np.sum(errors_mask & battery_mask) / np.sum(battery_mask) * 100
        battery_errors[battery] = battery_error_rate

# Plot top 10 batteries by error rate
top_batteries = sorted(battery_errors.items(), key=lambda x: x[1], reverse=True)[:10]
battery_names = [b[0] for b in top_batteries]
error_rates = [b[1] for b in top_batteries]
plt.bar(range(len(battery_names)), error_rates, color='red')
plt.xticks(range(len(battery_names)), battery_names, rotation=45)
plt.title('Top 10 Batteries by Error Rate')
plt.ylabel('Error Rate (%)')

# Plot 6: Error trend over sequence position
plt.subplot(2, 3, 6)
sequence_errors = []
for seq_pos in range(X_test_c.shape[1]):
    position_errors = np.sum(errors_mask) / len(errors_mask) * 100
    sequence_errors.append(position_errors)

plt.plot(range(X_test_c.shape[1]), sequence_errors, marker='o', color='red')
plt.title('Error Rate by Sequence Position')
plt.xlabel('Sequence Position')
plt.ylabel('Error Rate (%)')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("=" * 60)
print("‚úÖ ERROR ANALYSIS PLOTS COMPLETED!")
print("   These plots show WHERE and WHY the model makes errors")
```
